**Constructing a DataFrame using dict**

> csv 파일을 통하는 것이 아니라, 처음부터 만들기


*   dict는 key:value로 나뉨
*   value는 list로 이루어짐, 각 value들의 list의 item 수는 모두 같아야 함
*   dict의 한 item당 한 column이 됨




```python
# Some input data - a small sample of the iris dataset
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}
#dict의 item 5개, key:value로 나뉨

df = pd.DataFrame(data)
#pandas에 있는 dataframe 함수에 data를 입력으로 넣어주면 dataframe을 만들어준다 
df 
```



---



**Constructing a DataFrame using pd.Series**



> 한 column의 내용 부분만 series 함수로 따로
구성하고 dict로 만듬



```python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df
```



---



**Changing column names**


> column명을 수정할 때 ex. SPECIES -> species



```python
df.columns  #df.column 하면 위 표에서 column명만 나옴
```

```python
# Change the columns
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']  #list로 묶기
df
```



---



**Accesing columns**


*   특정 칼럼만 가지고 오고 싶을 때
*   df.loc[low, column]
*   항목 추가



```python
# Access one column 
df.loc[:, 'sepal_length']   #콜론(:)은 전부 다 -> 행(low)은 전부 다
```

```python
# Access one column (short-hand)
df['sepal_length']
```

```python
# Access one column (shortest-hand)
df.sepal_length
```

```python
# Access two columns
df.loc[:, ['sepal_length', 'sepal_width']]   #2개를 적을 때는 list로 묶기
```

```python
# Access two columns (short-hand)
df[['sepal_length', 'sepal_width']]
```



---



**Accessing rows**

```python
# Access one row 
df.loc[3, :]
```

```python
# Access two rows
df.loc[[3, 0], :]
```

```python
# Access a range of rows
df.loc[2:5, :]   #5는 빠지고 2, 3, 4
```

```python
# Filter keeping only rows with `sepal_length` > 5
df[df.sepal_length > 5]
```



---



**Accessing a Cell**

```python
df.loc[:, 'sepal_length'][4]    #sepal_length 통으로 가져왔다가 그 중 4 low
```

```python
df['sepal_length'][4]
```

```python
df.sepal_length[4]
```
